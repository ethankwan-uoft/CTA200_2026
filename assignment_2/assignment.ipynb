```python
# imports
import numpy as np
```

```python
# PART 1

def f(x):
    """
    This function calculates f(x) = x^3 - x^2 - 1.
    
    Parameters:
    x, array-like, the independent variable
    
    Returns:
    f, array-like, the dependent variable
    """
    
    return x**3 - x**2 - 1

def df(x):
    """
    This function calculates the derivative of f(x), df(x) = 3x^2 - 2x.
    
    Parameters:
    x, array-like, the independent variable
    
    Returns:
    f, array-like, the dependent variable
    """
    
    return 3*x**2 - 2*x
```

```python
# PART 2

def newton(f, df, x0, epsilon=1e-6, max_iter=30):
    """
    This function performs a Newton iteration on the function f to find a root.
    
    Parameters:
    f, function, the main function f
    df, function, the derivative of the function f
    x0, float, the initial guess of the root
    epsilon, float, the threshold in comparing a number y to 0, i.e. |y| < epsilon => y = 0
    max_iter, int, the maximum number of Newton iterations that should be performed
    
    Returns:
    x, float, the root of the function f
    """
    
    if df(x0) == 0:
        raise ValueError("Derivative at initial guess is 0, pick a different initial guess")
        
    n = 0
    x = x0
    while (n < max_iter) and (np.abs(f(x)) >= epsilon):
        x = x - f(x)/df(x)
        n = n + 1
    
    if np.abs(f(x)) < epsilon:
        print(f"Found root in {n} iterations")
        return x
    
    print("Iteration failed")
    return None
```

```python
newton(f, df, 2)
```

```python
newton(f, df, 1)
```

```python
newton(f, df, 2, epsilon=1e-8)
```

```python
newton(f, df, 1, epsilon=1e-8)
```
